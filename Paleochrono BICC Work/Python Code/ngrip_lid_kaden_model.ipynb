```python
import pandas as pd
import numpy as np

load = pd.read_excel('/Users/quinnmackay/Documents/GitHub/BICC/Data Storage/NGRIP/NGRIP_Kaden_Model.xlsx')
load.dropna(inplace=True)

load.rename(columns={'Age (yrbp)': 'age', 'LID (m)': 'LID'}, inplace=True)

gicc05 = pd.read_excel('/Users/quinnmackay/Documents/GitHub/BICC/Data Storage/Chronologies/gicc05_time_scale_excel.xlsx', skiprows=40)
```

```python
gicc05_ng = gicc05[['Age [a] (b2k)', 'Depth ice/snow [m] (NGRIP1)', 'Depth ice/snow [m] (NGRIP2)']].copy()

ng_combined = np.where(
    gicc05_ng['Depth ice/snow [m] (NGRIP1)'] < 1350,
    gicc05_ng['Depth ice/snow [m] (NGRIP1)'],
    gicc05_ng['Depth ice/snow [m] (NGRIP2)']
)

gicc05_ng['ng_combined'] = ng_combined

gicc05_ng.rename(columns={
    'Age [a] (b2k)': 'age',
    'Depth ice/snow [m] (NGRIP1)': 'depth_ng1',
    'Depth ice/snow [m] (NGRIP2)': 'depth_ng2',
}, inplace=True)

gicc05_ng_combined = gicc05_ng[['age', 'ng_combined']].copy()
load['depth'] = np.interp(load['age'], gicc05_ng_combined['age'], gicc05_ng_combined['ng_combined'])

output = load[['depth', 'LID']].copy()
output['rel_unc'] = 0.2
output['comment'] = np.nan

output.to_csv('/Users/quinnmackay/Desktop/lock_in_depth.txt', index=False, sep='\t')
```
