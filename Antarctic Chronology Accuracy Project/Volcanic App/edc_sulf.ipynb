```python
from pathlib import Path
import pandas as pd
import numpy as np
import os
```

```python
#Sulfate

edc_sulfate = pd.read_excel('/Users/quinnmackay/Documents/GitHub/BICC/Data Storage/Specific Sulfate/EDC/EDC_FIC_sulfate_7_3190m.xlsx', comment='#', sheet_name=1, usecols=[0,1], names=['depth', 'sulfate'])

edc_sulfate.dropna(inplace=True)
edc_sulfate.sort_values('depth', inplace=True)

file_name = 'edc_sulfate.txt'
core = 'edc'

#create dir if necessary
output_dir = f'/Users/quinnmackay/Documents/GitHub/BICC/Antarctic Chronology Accuracy Project/Plot_Sulfate/out/{core}'
os.makedirs(output_dir, exist_ok=True)

# save file
edc_sulfate.to_csv(f'{output_dir}/{file_name}', index=False)

```

```python
#DEP

edc_dep = pd.read_csv('/Users/quinnmackay/Documents/GitHub/BICC/Data Storage/Specific Sulfate/EDC/EDC96 DEP trimmed.txt', comment='#', sep=r'\s+', skiprows=55, names=['depth', 'dep'])

edc_dep.dropna(inplace=True)
edc_dep.sort_values('depth', inplace=True)

interp_load = pd.read_excel('/Users/quinnmackay/Documents/GitHub/BICC/Data Storage/Tiepoints/EDC96-EDC99.xlsx', skiprows=1)
edc_dep['edc99 depths'] = np.interp(edc_dep['depth'], interp_load['EDC96'], interp_load['EDC99'])

edc_dep_filtered = edc_dep[edc_dep['depth'] >= 12.31].copy(deep=True)

edc_dep_filtered = edc_dep_filtered[['edc99 depths', 'dep']]
edc_dep_filtered = edc_dep_filtered.rename(columns={'edc99 depths': 'depth'})

file_name = 'edc_dep.txt'
core = 'edc'

#create dir if necessary
output_dir = f'/Users/quinnmackay/Documents/GitHub/BICC/Antarctic Chronology Accuracy Project/Plot_Sulfate/out/{core}'
os.makedirs(output_dir, exist_ok=True)

# save file
edc_dep_filtered.to_csv(f'{output_dir}/{file_name}', index=False)

```
